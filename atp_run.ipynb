## 1. Load Code Cells Separately
Create one cell per logical block so you can verify each step before moving on.

```python
import os
import sys

sys.path.append(os.getcwd())
sys.path.append(os.path.join(os.getcwd(), "src"))

from atp_mlp_keras import (
    set_seeds,
    load_dataset,
    split_train_val_test,
    train_with_stratified_kfold,
    train_final_model,
 )
from ai_project.model_artifacts import (
    DEFAULT_METADATA_PATH,
    DEFAULT_MODEL_PATH,
    DEFAULT_SCALER_PATH,
    save_model_artifacts,
)

DATA_PATH = "atp_tennis.csv"
EXPORT_PDF = "outputs/confusion_matrix_notebook.pdf"
MODEL_PATH = DEFAULT_MODEL_PATH
SCALER_PATH = DEFAULT_SCALER_PATH
METADATA_PATH = DEFAULT_METADATA_PATH
```

```python
set_seeds()
x, y, feature_cols = load_dataset(DATA_PATH)
print("Feature columns:", feature_cols)
print("Total rows:", len(y))
```

```python
import pandas as pd

df_raw = pd.read_csv(DATA_PATH)

dup_mask = df_raw.duplicated()
dup_rows = int(dup_mask.sum())
dup_rows_all = int(df_raw.duplicated(keep=False).sum())

print("Duplicate rows (excluding first occurrence):", dup_rows)
print("Rows involved in duplicates:", dup_rows_all)
if dup_rows:
    print("Sample duplicate rows:")
    print(df_raw[dup_mask].head())
else:
    print("No duplicate rows found.")
```

## 2. Execute Cells One by One
Run the next cells in order to verify the split sizes and training steps.

```python
train_idx, val_idx, test_idx = split_train_val_test(y)
total = len(y)
print(f"Train: {len(train_idx)} ({len(train_idx) / total:.1%})")
print(f"Val:   {len(val_idx)} ({len(val_idx) / total:.1%})")
print(f"Test:  {len(test_idx)} ({len(test_idx) / total:.1%})")
```

## 3. Capture Output in Output Pane
The next cell runs training, prints the confusion matrix/metrics, and saves the model artifacts for Streamlit.

```python
EPOCHS = 50
BATCH_SIZE = 32

train_with_stratified_kfold(
    x,
    y,
    feature_cols,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    export_pdf=EXPORT_PDF,
)

final_model, final_scaler = train_final_model(
    x,
    y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
)
artifacts = save_model_artifacts(
    final_model,
    final_scaler,
    model_path=MODEL_PATH,
    scaler_path=SCALER_PATH,
    metadata_path=METADATA_PATH,
    metadata={
        "data_path": DATA_PATH,
        "feature_cols": feature_cols,
        "form_window": 10,
        "epochs": EPOCHS,
        "batch_size": BATCH_SIZE,
    },
)
print("Saved model:", artifacts.model_path)
print("Saved scaler:", artifacts.scaler_path)
print("Saved metadata:", artifacts.metadata_path)
```
